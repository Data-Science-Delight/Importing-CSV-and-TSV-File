## Open CSV File using URL:

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/Data-Science-Delight/titanic.csv/main/train%20(1).csv"

response = requests.get(url)
data = StringIO(response.text)

pd.read_csv(data)
```
